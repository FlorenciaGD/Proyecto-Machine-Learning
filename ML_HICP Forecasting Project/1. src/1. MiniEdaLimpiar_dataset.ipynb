## ESTRUCTURA DATASET
- freq: Meses
- unit: Unidad del índice (aquí "I05", que significa "índice, 2015 = 100").
- coicop: Classification of Individual Consumption According to Purpose, agrupar los artículos de la cesta de consumo utilizada en el cálculo del Índice de Precios al Consumidor (IPC) en subclases, clases, subgrupos y grupos, adaptándose completamente a la COICOP.(aquí "CP00", que representa el total general del índice)
- geo: Países
- date: Fechas
- hicp: Índice Armonizado de Precios de Consumo (Harmonised Index of Consumer Prices)

```python
# Filtrar entre dos fechas
start_date = '2018-01-01'
end_date = '2025-03-01'

df_date = df_long[(df_long['date'] >= start_date) & (df_long['date'] <= end_date)]
```

```python
coicop= ["CP00", "CP01", "CP011", "CP0111", "CP01111", "CP01112", "CP02", "CP03", "CP04", "CP06", "CP07"]
df_coicop= df_date[df_date["coicop"].isin(coicop)].copy()
```

```python
df_coicop
```

Los códigos coicop corresponden a una clasificación estándar internacional llamada COICOP (Classification of Individual Consumption by Purpose), utilizada para categorizar el gasto de los hogares en bienes y servicios. Se usa, por ejemplo, en los índices de precios al consumo (como el HICP).

CP00: Consumo total privado, que incluye todas las categorías de gastos de los hogares.

CP01 – Alimentos y bebidas no alcohólicas

CP011 – Alimentos

CP0111 – Pan y cereales

CP01111 – Pan

CP01112 – Cereales para desayuno

CP02 – Bebidas alcohólicas, tabaco y narcóticos

CP03 – Prendas de vestir y calzado

CP04 – Vivienda, agua, electricidad, gas y otros combustibles

CP05 – Mobiliario, equipamiento del hogar y mantenimiento corriente del hogar

CP06 – Salud

CP07 – Transporte

CP08 – Comunicaciones

CP09 – Ocio y cultura

CP10 – Educación

CP11 – Restaurantes y hoteles

CP12 – Bienes y servicios diversos

## También aparecen códigos personalizados o combinados como:

FOOD_P_X_ALC_TBC – Alimentos procesados excluyendo alcohol y tabaco

TOT_X_FUEL – Total del gasto sin combustibles

IGD_NNRG – Bienes industriales no energéticos

SERV_TRA – Servicios de transporte

## Agrupaciones especiales en "geo":
EA:	Zona Euro (genérico)

EA19:	Zona Euro con 19 países (hasta 2023) 

EA20:	Zona Euro con 20 países (desde 2023, incluye Croacia) 

EU:	Unión Europea (general) 

EU27_2020:	UE con 27 países tras salida del Reino Unido 

EU28:	UE con 28 países (incluía Reino Unido hasta 2020) 

EEA:	Espacio Económico Europeo (UE + Islandia, Liechtenstein y Noruega)
