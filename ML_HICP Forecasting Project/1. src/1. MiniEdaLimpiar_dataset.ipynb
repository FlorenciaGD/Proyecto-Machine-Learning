```python
df_price.columns = df_price.columns.str.strip()
df = df_price.rename(columns={'geo\\TIME_PERIOD': 'geo'})  # corrige el nombre de la columna geo si es necesario

df_long = df.melt(id_vars=['freq', 'unit', 'coicop', 'geo'],
                  var_name='date', value_name='hicp')

df_long['date'] = pd.to_datetime(df_long['date'], format='%Y-%m')
df_long['hicp'] = pd.to_numeric(df_long['hicp'], errors='coerce')
```

## ESTRUCTURA DATASET
- freq: Meses
- unit: Unidad del ndice (aqu "I05", que significa "ndice, 2015 = 100").
- coicop: Classification of Individual Consumption According to Purpose, agrupar los artculos de la cesta de consumo utilizada en el clculo del ndice de Precios al Consumidor (IPC) en subclases, clases, subgrupos y grupos, adaptndose completamente a la COICOP.(aqu "CP00", que representa el total general del ndice)
- geo: Pases
- date: Fechas
- hicp: ndice Armonizado de Precios de Consumo (Harmonised Index of Consumer Prices)

```python
# Filtrar entre dos fechas
start_date = '2018-01-01'
end_date = '2025-03-01'

df_date = df_long[(df_long['date'] >= start_date) & (df_long['date'] <= end_date)]
```

```python
coicop= ["CP00", "CP01", "CP011", "CP0111", "CP01111", "CP01112", "CP02", "CP03", "CP04", "CP06", "CP07"]
df_coicop= df_date[df_date["coicop"].isin(coicop)].copy()
```

```python
df_coicop
```

Los cdigos coicop corresponden a una clasificacin estndar internacional llamada COICOP (Classification of Individual Consumption by Purpose), utilizada para categorizar el gasto de los hogares en bienes y servicios. Se usa, por ejemplo, en los ndices de precios al consumo (como el HICP).

CP00: Consumo total privado, que incluye todas las categoras de gastos de los hogares.

CP01  Alimentos y bebidas no alcohlicas

CP011  Alimentos

CP0111  Pan y cereales

CP01111  Pan

CP01112  Cereales para desayuno

CP02  Bebidas alcohlicas, tabaco y narcticos

CP03  Prendas de vestir y calzado

CP04  Vivienda, agua, electricidad, gas y otros combustibles

CP05  Mobiliario, equipamiento del hogar y mantenimiento corriente del hogar

CP06  Salud

CP07  Transporte

CP08  Comunicaciones

CP09  Ocio y cultura

CP10  Educacin

CP11  Restaurantes y hoteles

CP12  Bienes y servicios diversos

## Tambin aparecen cdigos personalizados o combinados como:

FOOD_P_X_ALC_TBC  Alimentos procesados excluyendo alcohol y tabaco

TOT_X_FUEL  Total del gasto sin combustibles

IGD_NNRG  Bienes industriales no energticos

SERV_TRA  Servicios de transporte
